# An interactive notebook to design multiplex primers

## Load your table

```python
import pandas as pd

all_primers = pd.read_excel(r"Example_primer_sets.xlsx")
all_primers.head()
```

## Create instances of SaddleSimulator

```python
from saddle import SaddleSimulator

sim_1 = SaddleSimulator(
    all_primers = all_primers, # Your table.
    gt = 50, # Generation time. Default value is 100.
    len_substring = 4, # Length of a nucleotide substring. Default is 4.
    seed = 123 # Default is None but I suggest to write a random number.
)

sim_2 = SaddleSimulator(
    all_primers = all_primers, # Your table.
    gt = 50, # Generation time. Default value is 100.
    len_substring = 4, # Length of a nucleotide substring. Default is 4.
    seed = 9878 # Default is None but I suggest to write a random number.
)

sim_3 = SaddleSimulator(
    all_primers = all_primers, # Your table.
    gt = 50, # Generation time. Default value is 100.
    len_substring = 4, # Length of a nucleotide substring. Default is 4.
    seed = 456 # Default is None but I suggest to write a random number.
)
```

## Launch SADDLE algorithm

```python
last_set_of_primers_1 = sim_1.saddle()
last_set_of_primers_2 = sim_2.saddle()
last_set_of_primers_3 = sim_3.saddle()
```

```python
print("Set of primers #1:", last_set_of_primers_1, sep="\n")
print("Set of primers #2:", last_set_of_primers_2, sep="\n")
print("Set of primers #3:", last_set_of_primers_3, sep="\n")
```

```python
import matplotlib.pyplot as plt

cm = 1 / 2.54 # To use cm instead of inches for figure size

fig, ax = plt.subplots(1, 1, figsize = (16 * cm, 12 * cm))

sim_1.plot_loss(label = "sim_1", color = "blue")
sim_2.plot_loss(label = "sim_2", color = "red")
sim_3.plot_loss(label = "sim_3", color = "green")

plt.grid()
plt.minorticks_on()
plt.xlim(xmin = 0)
plt.ylim(ymin = 0)
plt.legend()
plt.xlabel("Generation")
plt.ylabel("Loss")
plt.title("Calculated loss at each generation")
plt.show()
```

This plot shows why we need to use SADDLE algorithm at least three times with set seeds. We can see that `sim_2` and `sim_3` found the best set of primers while `sim_1` failed to do so.
